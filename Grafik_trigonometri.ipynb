```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
t = np.arange(0,2,0.01)
s = np.sin(2 * np.pi * t)
t1 = t
s1 = s + 1
```

```python
# plot line
plt.figure(figsize=(8,5))
plt.plot(t,s,'bo--')
plt.plot(t1,s1)
plt.title('Grafik Sinus')
plt.xlabel('t')
plt.ylabel('nilai sinus')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.step(t,s,'yo--')
plt.title('Grafik Sinus Another Models')
plt.show()
```
